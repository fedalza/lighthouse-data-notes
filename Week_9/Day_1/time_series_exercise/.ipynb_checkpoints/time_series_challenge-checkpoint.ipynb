## Time Series Challenge
In this challenge, we will compare different time series prediction techniques on the dataset we can find [here](https://drive.google.com/file/d/1pgZlxluv5_cbgwR5JSWu4njbYXo_06Eb/view?usp=sharing).

There are several item categories in this sales data. You can choose your own category or predict overal sales.

> #### Note
> We have one order in every row, so don't forget to transform data into time series format first. The data can be transformed to daily, weekly or monthly timeseries.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Sample - Superstore.xls')
df.head(15)
```

```python
timeseries = df[['Order Date', 'Sales']]
timeseries = timeseries.sort_values(by=['Order Date'])
timeseries = timeseries.set_index(['Order Date'])
timeseries = timeseries.groupby(pd.Grouper(freq="w")).sum()
```

```python
timeseries.head()
```

```python
plt.plot(timeseries)
plt.show()
```

## Rolling mean

```python
rolling = timeseries.rolling(window=4)
rolling_mean = rolling.mean()

plt.plot(timeseries)
plt.plot(rolling_mean)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# prepare situation
X = timeseries.values
window = 4
history = [X[i] for i in range(window)]
test = [X[i] for i in range(window, len(X))]
predictions = list()
# walk forward over time steps in test
for t in range(len(test)):
    length = len(history)
    yhat = np.mean([history[i] for i in range(length-window,length)])
    obs = test[t]
    predictions.append(yhat)
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
```

## ARIMA

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARIMA
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

def ADF_Stationarity_Test(timeseries, significance_level=0.05):
    result = adfuller(timeseries)
    p = result[1]
    print('p-value: ', p)
    if p>significance_level:
        print('Reject H0, the time series is non-stationary')
    else:
        print('Cannot reject H0, the time series may be stationary')
```

```python
ADF_Stationarity_Test(timeseries,0.05)
```

```python
# evaluate combinations of p, d and q values for an ARIMA model
def evaluate_models(dataset, p_values, d_values, q_values):
    dataset = dataset.astype('float32')
    best_score, best_cfg = float("inf"), None
    for p in p_values:
        for d in d_values:
            for q in q_values:
                order = (p,d,q)
                try:
                    mse = evaluate_arima_model(dataset, order)
                    if mse < best_score:
                        best_score, best_cfg = mse, order
                    print('ARIMA%s MSE=%.3f' % (order,mse))
                except:
                    print('Didnt work', p, d, q)
                    continue
    print('Best ARIMA%s MSE=%.3f' % (best_cfg, best_score))

# evaluate an ARIMA model for a given order (p,d,q)
def evaluate_arima_model(X, arima_order):
    # prepare training dataset
    train_size = int(len(X) * 0.80)
    train, test = X[0:train_size], X[train_size:len(X)]
    history = []
    for x in train.Sales:
        history.append(x)
    #history.append([x for x in train.values])
    # make predictions
    predictions = list()
    for t in range(len(test)):
        model = ARIMA(history, order=arima_order)
        model_fit = model.fit(disp=0)
        yhat = model_fit.forecast()[0]
        predictions.append(yhat)
        #print(test,t)
        history.append(test.iloc[t])
    # calculate out of sample error
    error = mean_squared_error(test, predictions)
    return error

import warnings
warnings.filterwarnings("ignore")
```

```python
# evaluate parameters
p_values = [0, 1, 2, 4, 6, 8, 10]
d_values = range(0, 3)
q_values = range(0, 3)
warnings.filterwarnings("ignore")
evaluate_models(timeseries, p_values, d_values, q_values)
```

## LSTM

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler(feature_range = (0,1))
timeseries_sc = sc.fit_transform(timeseries)
```

```python
len(timeseries)*0.8
```

```python
X_train = []
y_train = []
for i in range(10,167):
    X_train.append(timeseries_sc[i-10:i,0])
    y_train.append(timeseries_sc[i,0])
X_train, y_train = np.array(X_train), np.array(y_train)

X_train = np.reshape(X_train,(X_train.shape[0], X_train.shape[1],1))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout

regressor = Sequential()

regressor.add(LSTM(units=50, return_sequences=True, input_shape = (X_train.shape[1],1)))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units=50))
regressor.add(Dropout(0.2))

regressor.add(Dense(units=1))

regressor.compile(optimizer='adam', loss='mean_squared_error')

regressor.fit(X_train,y_train, epochs = 100, batch_size=32)
```

```python
X_test = []
for i in range(167,len(timeseries)):
    X_test.append(timeseries_sc[i-10:i,0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1],1))
predicted = regressor.predict(X_test)
predicted = sc.inverse_transform(predicted)
```

```python
plt.plot(timeseries[167:], color = 'black')
plt.plot(timeseries.index[167:],predicted, color = 'green')
plt.show()
```
