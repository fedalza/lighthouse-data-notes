```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,100,200)
y = np.sin(x)

fig,ax = plt.subplots(figsize=(10,6))

ax.plot(x,y,"red")
plt.show()
```

```python
x = 10*np.random.rand(200,1)
y = 15*np.random.rand(200,1)

fig,ax = plt.subplots(figsize=(10,6))

ax.scatter(x,y)
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10,6))

ax.hist(y,bins=20)
plt.show()
```

```python
x = np.linspace(0,100,200)
y1 = np.sin(x)

y2 = (np.sin(x))*(np.cos(x))

fig,ax = plt.subplots(figsize=(10,6))

ax.plot(x,y1,'red')
ax.plot(x,y2,'blue')
plt.show()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(10,6))

axes[0].plot(x,y1,'red')
axes[0].set_title('sin(x)')

axes[1].plot(x,y2,'blue')
axes[1].set_title('sin(x)*cos(x)')

fig.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn import datasets
```

```python
iris = datasets.load_iris()

iris_df = pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df['target']=iris.target

iris_df['target_names']=iris_df.target.map(
    {0:'setosa',
    1:'versicolor',
    2:'virginica'}
)
```

```python
iris_df.head()
```

```python
setosa = iris_df[iris_df.target_names == 'setosa']

versicolor = iris_df[iris_df.target_names == 'versicolor']

virginica = iris_df[iris_df.target_names == 'virginica']

fig, ax = plt.subplots(figsize=(10,6))

ax.scatter(setosa['sepal length (cm)'],setosa['sepal width (cm)'],
           marker = 'o',color = 'red', label = 'setosa')
ax.scatter(versicolor['sepal length (cm)'],versicolor['sepal width (cm)'],
           marker = 'o',color = 'green', label = 'versicolor')
ax.scatter(virginica['sepal length (cm)'],virginica['sepal width (cm)'],
           marker = 'o',color = 'blue', label = 'virginica')

ax.legend(loc='upper right')
ax.set_title('Iris flower')
ax.set_xlabel('sepal length (cm)')
ax.set_ylabel('sepatl width (cm)')

plt.show()
```
