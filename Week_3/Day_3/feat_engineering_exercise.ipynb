```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import datetime
```

```python
df_train=pd.read_csv('UFO_data.csv',sep=';',index_col='Unnamed: 0')
```

```python
df_train.head()
```

```python
df_train.dtypes
```

```python
num_feats = df_train.dtypes[df_train.dtypes != 'object'].index.tolist()
df_numeric = df_train[num_feats]
df_numeric.dtypes
```

```python
df_numeric['days_to_report']=(df_train['date_reported'].apply(lambda x: datetime.datetime.strptime(x,'%Y-%m-%d %H:%M:%S').date())-df_train['date_event'].apply(lambda x: datetime.datetime.strptime(x,'%Y-%m-%d').date())).dt.days
```

```python
df_numeric['days_to_report']
```

```python
df_numeric.head()
```

```python
# TIME OF DAY -> MORNING (6-12), AFTERNOON(12-18), EVENING(18-24), NIGHT(0-6)
df_train['timestamp'][0]
df_train['period_of_day'] = 0
for i in range(len(df_train)):
    try:
        tyme = datetime.datetime.strptime(str(df_train.loc[i,'timestamp']),'%Y-%m-%d %H:%M:%S').hour
        if (tyme>=0) & (tyme<6):
            df_train.loc[i,'period_of_day']=0
        elif (tyme>=6) & (tyme<12):
            df_train.loc[i,'period_of_day']=1
        elif (tyme>=12) & (tyme<18):
            df_train.loc[i,'period_of_day']=2
        elif (tyme>=18) & (tyme<=23):
            df_train.loc[i,'period_of_day']=3
    except:
        df_train.loc[i,'period_of_day']=np.nan
        
        #df_train['timestamp'].apply(lambda x: datetime.datetime.strptime(str(x),'%Y-%m-%d %H:%M:%S'))
```

```python
df_numeric['period_of_day'] = df_train['period_of_day']
```
